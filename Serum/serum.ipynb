If we consider A == 440 Hz:

|Note |Frequency (Hz)  | Ratio floored (Hz/440)
|:--- | ---: | ---:|
|A| 440.00|1|
|C#| 554.37|1.25|
|E| 659.26|1.50|
|A2| 880.00|2|

````python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('dark_background')

# Visual representation of A, C#, E -- or three sine waves at 1, 1.25 and 1.50

def wave(ratio):
    cycles = 8.03 # .03 for margin when plotting visuals
    resolution = 256 # datapoints to generate

    length = np.pi * ratio * cycles

    plt.xlim(-10,resolution+10)
    plt.ylim(-1.25,1.25)

    return np.sin(np.arange(0, length, length / resolution))


fig, ax = plt.subplots()
plt.rc('figure', figsize=(16, 9), dpi=200)

for i in [1,1.25,1.5]:
    ax.plot(wave(i))

ax.axes.get_xaxis().set_visible(False)
plt.grid(color='#333333', linestyle='--')
plt.show()
````

This would represent a chord as three separate waves. In this representation, every sine starts of at the same time and eventually phases back into its origin.

## Briefly about frequencies of sound

Anything below 20 Hz falls into [infrasound](https://en.wikipedia.org/wiki/Infrasound), which are frequencies below human hearing.<br>
Everything above 20 kHz we label [ultrasound](https://en.wikipedia.org/wiki/Ultrasound), which are frequencies above humen hearing.

This gives us an audible spectra which supposedly fit everything we perceive as audio.<br>
For convenience, audio engineers and musicians often labels ranges of frequencies:

|Freq (Hz) |Domain  |
|:--- | ---: |
|<60| Sub|
|60-250| Bass|
|250-8k| Mid|
|8k>| Highs|
